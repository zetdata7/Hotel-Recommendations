# EXPORTING CLICKED AND BOOKED HOTELS BY USER
# OUT OF ATTRIBUTE: 'IMPRESSIONS'

 The following function is accepting as
input a string and splits it on char '|'
If the strings that were designed after the split have as last 3 characters these--> '0,0'
it means that this string's hotel was neither clicked nor booked. I do this in order to keep
the rest of the hotels. So the function return the strings with hotels booked or clicked

```python

def func(x):
    b = []
    x = x.split('|')
    for i in x:
        if i[-3:] != '0,0':
            b.append(i)
    if len(b) == 0:
        b = 0
    return b
```

The nan function aims to replace the strings named 'null' to actual null value.

```python
def nan(x):
    for i in range(0,len(x)):
        for j in range(0,len(x[i])):
            if x[i][j] =='null':
                x[i][j] = float("NaN")

    return x
```

```python
df['impressions']=df['impressions'].apply(nan)
```

The num function transforms the strings of the array to float.

```python
def num(x):
    x = x.astype('float64')
    return x
```

```python
df['impressions']=df['impressions'].apply(num)
```

Isolating only the hotel IDs which are the second element of each array.

```python
def hotel_id(x):
    a=[]
    for i in range (0,len(x)):
        a.append(x[i][1])
    return a
```

```python
df['hotels'] = df['list_of_impressions'].apply(hotel_id)
```

```python
def stringagain(x):
    x = [int(i) for i in x]
    x = [str(i) for i in x]
    return x
```

```python
w = w.apply(stringagain)
```

```python
def duplhotels(x):
    a=[]
    for i in x:
        if i not in a:
            a.append(i)
    return a
```

Above we used stringagain which transforms the hotel IDs to strings and duplhotels which deletes the duplicates hotels

```python
w = w.apply(duplhotels)
```

TUNING THE EMBEDDINGS\
Below we saved many models which will be evaluated on the 'Model Selection_Final.ipynb'

```python
k = 3
q = 5
for i in range(1,3):
    for l in range(1,5,1):
        mymodel = Word2Vec(X_train,min_count = i,vector_size = 32,window = l,negative = q, workers = k,sg=1,hs = 1)
        mymodel.save(f"hotel2vec,min{i}neg{q}work{k}window{l}skiphs.model")
```
